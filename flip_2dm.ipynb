```python
import pandas as pd
import numpy as np

fp = "projects/kysuce/tmp/003.2dm"

df = pd.read_csv(fp, skiprows=17, sep="\t")
df = df.loc[df.iloc[:, 8] != 0]

el_max = max(df.iloc[:, [0, 1, 2, 3]].max())

electrodes_reference = pd.Series(np.arange(el_max, -1, step=-1), name="new")

electrodes_new = []
for i in [1, 0, 3, 2]:
    tmp = df.copy().iloc[:, i]
    tmp.name = "old"

    tmp = pd.merge(
        tmp, electrodes_reference, how="left", left_on="old", right_index=True
    )
    electrodes_new.append(tmp["new"].values)

electrodes_new = np.column_stack(electrodes_new)


df.iloc[:, :4] = electrodes_new

df.to_clipboard(index=False, header=None)
```
