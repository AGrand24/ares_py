```python
import geopandas as gpd
import pandas as pd
import numpy as np

electrodes = [
    [-3, -1, 1, 3],
    [-9, -3, 3, 9],
    [-15, -5, 5, 15],
    [-29.4, -9, 9, 29.4],
]
electrodes = [
    [-1.5, -0.5, 0.5, 1.5],
    [-5, -1.5, 1.5, 5],
    [-7.5, -2.5, 2.5, 7.5],
    [-15, -5, 5, 15],
]
electrodes = np.array(electrodes)
l = np.ptp(electrodes, axis=1)
d = l * 0.173
a = electrodes[:, 2] - electrodes[:, 1]
n = (electrodes[:, 1] - electrodes[:, 0]) / a
k = np.pi * n * (n + 1) * a


array = np.column_stack([electrodes, l, d, a, n, k])
array = np.round(array, 1)

df = pd.DataFrame(array)
df.columns = ["C1", "P1", "P2", "C2", "l", "doi", "a", "n", "k"]
fp = "projects/kysuce/array_kysuce.gpkg"

gpd.GeoDataFrame(df).to_file(fp, layer="tab")

x = electrodes.flatten()
y = np.arange(0, (len(x)))
y = y // 4
y *= -4


geom = gpd.points_from_xy(x, y)
gdf = gpd.GeoDataFrame(geometry=geom, crs=8353)
gdf[["x", "y"]] = gdf.get_coordinates()
gdf.to_file(fp)

el_pos = np.unique(electrodes)
x = el_pos
y = -16
geom = gpd.points_from_xy(x, y)
gdf = gpd.GeoDataFrame(geometry=geom, crs=8353)
gdf[["x", "y"]] = gdf.get_coordinates()
gdf.to_file(fp, layer="electrodes")
```
